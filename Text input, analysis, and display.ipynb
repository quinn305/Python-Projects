# Transcript

```python
num_of_segments = 10
def Transcript(segs):
    words = []
    for i in range(segs):
        data = input("Please transcribe: ")
        words.append(data)
    return words
#transcript = Transcript(num_of_segments)
#print(f'The transcript is: {transcript}')
```

```python
def transcriptWords(wordlist):
    wordlist2 = []
    for i in wordlist:
        word = i.split()
        wordlist2.append(word)
    
    wordlist3 = []
    for j in wordlist2:
        for k in j:
            wordlist3.append(k)
    
    return wordlist3

#words_of_transcript = transcriptWords(transcript)
#print(words_of_transcript)
```

# Phrasal Translation

```python
def phraseTranslate(wordlist):
    phrasal_translation_list = []
    for i in wordlist:
        data = input(f"Translate '{i}' phrasally: ")
        phrasal_translation_list.append(data)
    
    return phrasal_translation_list

#phrasal_translation = phraseTranslate(transcript)
#print(phrasal_translation)
```
