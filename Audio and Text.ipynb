```python
import pyaudio
import wave
from mutagen.wave import WAVE
import math
```

```python
def getTime(audio):
    real_time = audio.info.length
    real_time_ceil = math.ceil(real_time)
    print(f"The audio file is {real_time_ceil}s long")
```

```python
def audioSlices(file_name, slice_length):
    counter = 20
    while counter <= real_time_ceil:
        # set desired values
        start = counter
        length = slice_length
        end = start + length
        print(f"Playing {start}s to {end}s")

        # open wave file
        wave_file = wave.open(file_name, 'rb')

        # initialize audio
        py_audio = pyaudio.PyAudio()
        stream = py_audio.open(format=py_audio.get_format_from_width(wave_file.getsampwidth()),
                               channels=wave_file.getnchannels(),
                               rate=wave_file.getframerate(),
                               output=True)

        # skip unwanted frames
        n_frames = int(start * wave_file.getframerate())
        wave_file.setpos(n_frames)

        # write desired frames to audio buffer
        n_frames = int(length * wave_file.getframerate())
        frames = wave_file.readframes(n_frames)
        stream.write(frames)

        # close and terminate everything properly
        stream.close()
        py_audio.terminate()
        wave_file.close()
        test = input("Ready for the next slice? (yes/redo)")
        print(test)
        if test == 'yes':
            counter += length
        else:
            print('Audio slice is being replayed')
```
