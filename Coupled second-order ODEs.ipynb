```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import math 
 
# 1 parameter omega
om = 1
 
# defining the set of four, coupled 1st-order odes, after converting from two, coupled 2nd-order odes.
def model(sol, t):
    x1, y1, x2, y2 = sol
 
    dx1dt = x2
    dx2dt = (om**2)*x1 + 2*om*y2
    dy1dt = y2
    dy2dt = (om**2)*y1 - 2*om*x2
 
    return [dx1dt, dy1dt, dx2dt, dy2dt]
 
# creating a time domain
t_span = np.linspace(0, 10, 100)
 
# ICs
x1_0 = 1
x2_0 = 0
y1_0 = 1
y2_0 = 0
 
sol_zero = [x1_0, y1_0, x2_0, y2_0]
 
solution = odeint(model, sol_zero, t_span)

#plot trajectory y/x
plt.plot(solution[:, 0], solution[:, 1], label='y/x')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
```

```python
#r(t)
plt.plot(t_span, solution[:, 0], label='x(t)')
plt.plot(t_span, solution[:, 1], label='y(t)')
plt.legend()
plt.xlabel('t')
plt.ylabel('position')
```

```python
#V(t)
plt.plot(t_span, solution[:, 2], label='Vx(t)')
plt.plot(t_span, solution[:, 3], label='Vy(t)')
plt.legend()
plt.xlabel('t')
plt.ylabel('velocity')
```

```python
#Kinetic Energy
m = 1
Tx = 0.5*m*(solution[:, 2])**2
Ty = 0.5*m*(solution[:, 3])**2
plt.plot(t_span, Tx, label='Tx')
plt.plot(t_span, Ty, label='Ty')
```

```python
# Potential Energy
g = 9.81
h = solution[:, 1]
U = -m*g*h
plt.plot(t_span, U, label='U')
```

```python
Ey = Ty + U
plt.plot(t_span, Ey, label='Ey')
```
