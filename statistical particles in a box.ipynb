# 8 particles in a box

```python
import math
import numpy as np

configurations = [1,2,3,4,5,6,7,8,9]
print(f'List of configurations: {configurations}')
print("")

n1 = [8-i for i in range(9)]
print(f'list of n1: {n1}')
print("")

n2 = [0+i for i in range(9)]
print(f'list of n2: {n2}')
print("")

trues1 = 0
for i,j in zip(n1,n2):
    if (i or j) < 3 or (i or j) >5:
        trues1+=1
print(f'Probability of being away from equilibrium:',trues1/9)
print("")

W = [math.factorial(8)/(math.factorial(i)*math.factorial(j)) for i, j in zip(n1,n2)]
print(f'list of W: {W}')
print("")

S = [1.38*np.log(i) for i in W]
print(f'list of S: {S}')
print("")

lst8 = [[a,b,c,d,e] for a,b,c,d,e in zip(configurations,n1,n2,W,S)]
print("Spreadsheet:")
lst8_head = ['configurations','n1','n2','W','S']
print(lst8_head)
for row in lst8:
    print(row)
```

```python
import matplotlib.pyplot as plt
print('For the 8 particle configurations')
plt.plot(W,n1)
plt.ylabel('n1')
plt.xlabel('W')
plt.show()

plt.plot(S,n1)
plt.ylabel('n1')
plt.xlabel('S')
plt.show()
```

# 50 particles in a box

```python
configurations_50 = [i for i in range(1,51)]
print(f'Configuration list is: {configurations_50}')
print("")

n1_50 = [50-i for i in range(51)]
print(f'n1 list is: {n1_50}')
print("")

n2_50 = [0+i for i in range(51)]
print(f'list of n2: {n2_50}')
print("")

trues50 = 0
for i,j in zip(n1,n2):
    if (i or j) < 20 or (i or j) >30:
        trues50+=1
print(f'Probability of being away from equilibrium is:',trues50/51)
print("")

W_50 = [math.factorial(50)/(math.factorial(i)*math.factorial(j)) for i, j in zip(n1_50,n2_50)]
print(f'list of W: {W_50}')
print("")

S_50 = [1.38*np.log(i) for i in W_50]
print(f'list of S: {S_50}')
print("")

lst = [[a,b,c,d,e] for a,b,c,d,e in zip(configurations_50,n1_50,n2_50,W_50,S_50)]
print("Spreadsheet:")
lst_head = ['configurations','n1','n2','W','S']
print(lst_head)
for row in lst:
    print(row)
```

```python
print('For the 50 particle configurations')
plt.plot(W_50,n1_50)
plt.ylabel('n1')
plt.xlabel('W')
plt.show()

plt.plot(S_50,n1_50)
plt.ylabel('n1')
plt.xlabel('S')
plt.show()
```

The probability of being far from equilibrium decreases exponentially, as observed by comparing the probabilities of being away from equilibrium from the 8 particle and 50 particle examples.

# After Refactoring

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def statistics(num, l,h):

    configurations = [i for i in range(num+2)]
    print(f'List of configurations: {configurations}')
    print("")

    n1 = [num+1-i for i in range(num+2)]
    print(f'list of n1: {n1}')
    print("")
    
    n2 = [0+i for i in range(num+2)]
    print(f'list of n2: {n2}')
    print("")

    trues1 = 0
    for i,j in zip(n1,n2):
        if (i or j) < l or (i or j) >h:
            trues1+=1
    print(f'Probability of being away from equilibrium:',trues1/(num+1))
    print("")

    W = [math.factorial(num)/(math.factorial(i)*math.factorial(j)) for i, j in zip(n1,n2)]
    print(f'list of W: {W}')
    print("")

    S = [1.38*np.log(i) for i in W]
    print(f'list of S: {S}')
    print("")

    lst8 = [[a,b,c,d,e] for a,b,c,d,e in zip(configurations,n1,n2,W,S)]
    print("Spreadsheet:")
    lst8_head = ['configurations','n1','n2','W','S']
    print(lst8_head)
    for row in lst8:
        print(row)


    
    print('For the 8 particle configurations')
    plt.plot(W,n1)
    plt.ylabel('n1')
    plt.xlabel('W')
    plt.show()

    plt.plot(S,n1)
    plt.ylabel('n1')
    plt.xlabel('S')
    plt.show()
```

```python
statistics(8,3,5)
```
